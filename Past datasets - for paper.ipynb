# Wiser et al, 2013 dataset

```python
def get_freqs(arr):
    times = arr[:,0]
    ycurr0 = arr[:,1]
    yanc0 = arr[:,2]
    ycurrf = arr[:,3]
    yancf = arr[:,4]

    f_f = ycurrf/(ycurrf + yancf)
    f_i = ycurr0/(ycurr0 + yanc0)
    return times, f_i,f_f

def compute_fitness(f_i,f_f):
    return 100*np.log(f_f*(1-f_i)/(f_i*(1-f_f)))/6.7
```

```python
plt.rc('axes', prop_cycle=(cycler('color', palettable.colorbrewer.qualitative.Dark2_3.mpl_colors)))

plt.close("all")
data0 = deepcopy(data_mean[:,:,0])
data0 = data0[data0 == data0]

data1 = deepcopy(data_mean[:,:,1])
data1 = data1[data1 == data1]

data2 = deepcopy(data_mean[:,:,2])
data2 = data2[data2 == data2]

print(data0.shape, data1.shape, data2.shape)

plt.hist(data0)
plt.hist(data1 + 5)
plt.hist(data2 + 10)
plt.show()
```

```python
#Plot data

plt.rc('axes', prop_cycle=(cycler('color', palettable.colorbrewer.qualitative.Dark2_3.mpl_colors)))

diffs1 = data_mean[:,:,1] - data_mean[:,:,0]
diffs2 = data_mean[:,:,2] - data_mean[:,:,0]

diffs1_mean = np.nanmean(diffs1,axis=1)
diffs2_mean = np.nanmean(diffs2,axis=1)

diffs1_ste = np.nanstd(diffs1,axis=1)/np.sqrt(10)
diffs2_ste = np.nanstd(diffs2,axis=1)/np.sqrt(10)

data_mean_mean0 = np.nanmean(data_mean[:,:,0],axis=1)
data_mean_ste0 = np.nanstd(data_mean[:,:,0],axis=1)/np.sqrt(10)

plt.close("all")
fig,axis = plt.subplots(1,1,figsize = (4.5,4))

axis.errorbar(data_mean_mean0, diffs1_mean, xerr = data_mean_ste0, yerr = diffs1_ste, fmt = 'C0o',ms = 4, label = "Gen 250")
axis.errorbar(data_mean_mean0, diffs2_mean, xerr = data_mean_ste0, yerr = diffs2_ste, fmt = 'C1o',ms = 4, label = "Gen 500")

fit1 = np.polyfit(data_mean_mean0, diffs1_mean,1)
fit2 = np.polyfit(data_mean_mean0, diffs2_mean,1)

xs = np.linspace(-10,10,100)
axis.plot(xs, xs*fit1[0] + fit1[1], 'C0--' ,lw = 2)
axis.plot(xs, xs*fit2[0] + fit2[1], 'C1--' ,lw = 2)

axis.tick_params(labelsize = 18, width = 2, size = 5)

axis.legend(fontsize = 18, frameon = False )
axis.set_xlabel(r"Initial relative fitness (%)",fontsize = 20)
axis.set_ylabel("Fitness gain (%)",fontsize = 20)

axis.set_yticks([0,5,10,15])
axis.set_xticks([-10,-5,0,5,10])
axis.set_ylim(-2,13)
axis.set_xlim(-10,8)
axis.spines['top'].set_linewidth(0)
axis.spines['left'].set_linewidth(1.5)
axis.spines['bottom'].set_linewidth(1.5)
axis.spines['right'].set_linewidth(0)
fig.tight_layout()
fig.savefig("./RandomQTL_Plots/krya_fig1_data.pdf")
plt.show()
```

# Johnson et al 2019

```python
#Load data

df = pd.read_excel("science_data_clean.xlsx")
df.as_matrix()

data_arr = np.array(df)

nz_cols = (data_arr[-1,:] == data_arr[-1,:])*(np.nanmean(data_arr[:-1],axis=0) == np.nanmean(data_arr[:-1],axis=0))

dfe = 100*data_arr[:-1,nz_cols]
brg = 100*data_arr[-1,nz_cols]

dfe_mean = np.nanmean(dfe,axis=0)
dfe_std = np.nanstd(dfe,axis=0)
#dfe_skew = ss.skew(dfe,axis=0)
```

```python
plt.close("all")
cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
#plt.rc('axes', prop_cycle=(cycler('color', palettable.cubehelix.cubehelix3_16.mpl_colors)))
plt.rc('axes', prop_cycle=(cycler('color', palettable.colorbrewer.qualitative.Dark2_5.mpl_colors)))
#Indices sorted by \hat{v}
# array([38, 20, 79, 26, 12, 61, 62, 52,  3, 35, 42, 19, 46, 56, 68, 87, 94,
#        13, 69, 21, 73, 93, 53, 63, 36, 76,  7, 70, 90, 27, 11, 65, 59, 45,
#        80, 74, 49, 15, 81, 55, 23, 91, 31, 39, 29, 50, 71, 24, 17, 92, 60,
#        51, 30, 18, 58, 41,  1, 43, 97, 32, 40, 28,  2, 85,  9, 34, 48, 72,
#        98,  5, 67, 64, 10, 66,  8, 75,  4,  6, 54, 33, 57, 14, 44, 22, 96,
#        88, 84, 89, 37, 25, 86, 78, 47])

ranges = np.arange(-0.1,0.1,0.025)*100

dfe_filts = []
brg_filts = []
dfe_filts_ste = []
brg_filts_ste = []

inds = [38, 61, 35, 2, 4] 
dfe_inds_filts = []
dfe_inds_filts_ste = []
for i in inds:
    dfe_inds_filts += [[]]
    dfe_inds_filts_ste += [[]]

for i in range(len(ranges)-1):
    filt = (brg > ranges[i])*(brg < ranges[i+1])
    dfe_filt = dfe[:,filt]
    dfe_filt = dfe_filt.flatten()
    dfe_filt = dfe_filt[dfe_filt == dfe_filt]
    
    dfe_filts += [np.mean(dfe_filt)]
    brg_filts += [np.mean(brg[filt])]
    
    dfe_filts_ste += [np.std(dfe_filt)/np.sqrt(len(dfe_filt))]
    brg_filts_ste += [np.std(brg[filt])/np.sqrt(len(brg[filt]))]
    
    for j,ind in enumerate(inds):
        dfe_ind_filt = dfe[ind][filt]
        dfe_ind_filt = dfe_ind_filt[dfe_ind_filt == dfe_ind_filt]
        dfe_inds_filts[j] += [np.nanmean(dfe_ind_filt)]
        dfe_inds_filts_ste[j] += [np.nanstd(dfe_ind_filt)/np.sqrt(len(dfe_ind_filt))]

fig,axis = plt.subplots(1,1,figsize = (4.5,4))
fac = 1
for j in range(len(inds)):
    color = 1*j
    fits = np.polyfit(brg_filts,dfe_inds_filts[j]- fac*dfe_inds_filts[j][0],1)
    x = np.linspace(-0.1,0.1,10)*100
    axis.plot(x,x*fits[0] + fits[1],'C%d--'%(color),lw = 2, alpha = 1)
    axis.errorbar(brg_filts,dfe_inds_filts[j] - fac*dfe_inds_filts[j][0], xerr  = brg_filts_ste, yerr = dfe_inds_filts_ste[j], fmt = 'C%do'%(color), ms = 6, alpha = 1)

fits = np.polyfit(brg_filts,dfe_filts - fac*dfe_filts[0],1)
x = np.linspace(-0.1,0.1,10)*100
axis.plot(x,x*fits[0] + fits[1],'-',color = 'r',lw = 2,alpha = 1)

axis.errorbar(brg_filts,dfe_filts - fac*dfe_filts[0], xerr  = brg_filts_ste, yerr = dfe_filts_ste, fmt = 'o',color ='r', ms = 7, alpha = 1, label = "Mean (91 loci)")

    
    
#axis.set_xlim(-.13,0.12)
#axis.set_ylim(-0.05,0.01)
axis.set_xlabel(r"Background fitness (%)",fontsize = 20)
axis.set_ylabel("Relative fitness effect (%)",fontsize = 20)
axis.tick_params(labelsize = 18, width = 2, size = 5)
leg = axis.legend(fontsize = 18, frameon = False)

axis.set_yticks([-10,0])
axis.set_xticks([-8,0,8])
axis.set_ylim(-16,1)
axis.set_xlim(-10,8)
axis.spines['top'].set_linewidth(0)
axis.spines['left'].set_linewidth(1.5)
axis.spines['bottom'].set_linewidth(1.5)
axis.spines['right'].set_linewidth(0)
fig.tight_layout()
fig.savefig("./RandomQTL_Plots/johnsonetal_binned_dfemean.pdf")
plt.show()
```
